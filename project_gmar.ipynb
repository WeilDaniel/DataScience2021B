```python

import plotly.offline as py
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.offline as offline
import plotly.graph_objs as go
```

```python
death = pd.read_csv('causes of death.csv')
death
```

```python
country = death.groupby('Entity')[['Terrorism']].sum()
country
```

```python
data = [ dict(
        type = 'choropleth',
        locations = death['Code'],
        z = country['Terrorism'],
        text = death['Entity'],
        colorscale = [[0,"rgb(255, 60, 20)"],[0.25,"rgb(255, 100, 20)"],[0.5,"rgb(255, 140, 20)"],[0.75,"rgb(255, 200, 20)"],[1,"rgb(225, 255, 150)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(240, 240, 240)',
                width = 1.5 ) ),
        colorbar = dict(
            autotick = False,
            title = 'terorisem by country'),
      ) ]

layout = dict(
    title = 'terrorism',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
    
)
fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='causes of death.csv')
```
